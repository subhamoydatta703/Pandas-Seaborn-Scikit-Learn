### The `seaborn.countplot()` function is used to show the counts of observations in each categorical bin using bars. It works like a histogram across a categorical variable instead of a quantitative one.

syntax:\
import seaborn as sns\
import matplotlib.pyplot as plt

sns.countplot(data=df, x=target_values_on_x, y=target_values_on_y, hue=None, palette=None, order=None)

Essential Parameters:\
data: The dataset (DataFrame, array, or list) used for plotting.\
x or y: The categorical variable column name.\
Setting x creates vertical bars; setting y creates horizontal bars.\
hue: Splits the bars based on an additional categorical subgroup.\
palette: Changes the color scheme of the bars (e.g., 'Set2', 'viridis').\
order: Accepts a list of strings to arrange the order of the categories on the axis


```python
sns.countplot(data=df, x="survived")
plt.show()
```
