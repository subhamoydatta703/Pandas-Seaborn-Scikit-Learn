### What are Bins?

Bins divide a numerical feature into intervals (ranges).

A histogram counts how many values fall inside each interval and displays them as bars.

Example:

Age values:
18, 20, 22, 25, 28, 35, 40, 45

With `bins=4`:

- 15–22
- 22–29
- 29–36
- 36–45

More bins → more bars → more detailed view of the distribution.

Fewer bins → fewer bars → a more general overview.

**Important:** More bins do NOT change the actual x-axis range or the underlying data values. They only divide the same x-axis range into smaller intervals.

Example:

Age range = 0–80

- `bins=5` → 5 wide bars across 0–80
- `bins=20` → 20 narrower bars across 0–80

The x-axis still represents 0–80 in both cases.

### Changing X-Axis Values (Ticks)

Bins control the bars, not the numbers shown on the x-axis.

To display more x-axis values:

## Boxplot

A boxplot is used to understand the spread of numerical data and to identify unusual values (outliers).

Example:

```python
sns.boxplot(data=df, x="age")
plt.show()
```

A boxplot gives a quick summary of a feature by showing:

- Where most values are concentrated
- The typical value (median)
- Values that are far away from the majority of the data

### Why ML Engineers Use Boxplots

Before training a machine learning model, we need to understand the quality of the data.

A boxplot helps answer questions like:

- Does this feature contain unusual values?
- Are there possible data entry mistakes?
- Should I investigate some observations before training the model?

For example, if most passenger ages are between 20 and 40, but a few passengers are 70-80 years old, a boxplot will highlight those values.

### Understanding Outliers

Outliers are values that are very different from the majority of the data.

However:

> An outlier does not automatically mean the data is wrong.

Examples:

- Age = 80 → Unusual, but realistic
- Age = 800 → Unusual and unrealistic

The first value may be kept, while the second should be investigated.

### How to Handle Outliers

When a boxplot shows unusual values:

1. Identify the outlier.
2. Check whether the value is realistic.
3. Understand the dataset and domain.
4. Decide what action to take.

Possible actions:

- Keep the value if it is a valid observation.
- Fix the value if it is a data entry error.
- Remove the value if it is invalid and cannot be corrected.

### Titanic Example

Suppose a boxplot of the `age` feature shows a few passengers with ages around 70-80 years.

These ages appear as outliers because most passengers are younger.

As an ML engineer, we would usually keep these rows because:

- The values are realistic.
- Elderly passengers can exist.
- The model should learn from real-world data.

### Key Takeaway

A boxplot is an **outlier detection tool**, not an **outlier removal tool**.

Its purpose is to help us find unusual values and investigate them before training a machine learning model. The final decision should always be based on understanding the data, not just the plot.

## Hue

`hue` is used to split data into categories and display each category with a different color.

The main purpose of `hue` is not decoration, but comparison.

### Why Use Hue?

Without `hue`, all data is shown together.

```python
sns.histplot(data=df, x="age")
```

This shows the age distribution of all passengers combined.

With `hue`:

```python
sns.histplot(data=df, x="age", hue="survived")
```

The data is split into groups:

- Survived (1)
- Died (0)

Each group gets a different color, making comparison easier.

### Example

```python
sns.countplot(data=df, x="pclass", hue="survived")
```

This answers:

> For each passenger class, how many passengers survived and how many died?

The bars are colored based on the values in the `survived` column.

### How Hue Works

```python
hue="survived"
```

Values:

```text
0 = Died
1 = Survived
```

Seaborn assigns a different color to each value.

---

```python
hue="sex"
```

Values:

```text
male
female
```

Seaborn assigns a different color to each category.

### Best Practice

`hue` works best with categorical features that have a small number of unique values, such as:

- survived
- sex
- pclass
- species

Using `hue` with features that have too many unique values can make plots difficult to read.

### Key Takeaway

> Hue is used to separate data into groups and color each group differently so that comparisons become easier during EDA.

# Correlation and Heatmap

## What is Correlation?

Correlation measures how strongly two numerical features are related.

It helps us understand whether two variables tend to move together.

### Correlation Scale

```text
+1 → Strong Positive Relationship

 0 → No Relationship

-1 → Strong Negative Relationship
```

### Positive Correlation

When one feature increases, the other feature also tends to increase.

Example:

```text
Study Hours ↑
Score ↑
```

Correlation:

```text
Close to +1
```

---

### Negative Correlation

When one feature increases, the other feature tends to decrease.

Example:

```text
Speed ↑
Travel Time ↓
```

Correlation:

```text
Close to -1
```

---

### No Correlation

When changes in one feature do not affect the other feature.

Example:

```text
Shoe Size ↔ Exam Score
```

Correlation:

```text
Close to 0
```

---

## Why ML Engineers Use Correlation

Before training a model, we want to know:

* Which features are related to the target?
* Which features may be useful for prediction?
* Are some features carrying similar information?
* Are there strong relationships between features?

Correlation helps answer these questions quickly.

# Heatmap

A heatmap is a visual representation of a correlation matrix.

Instead of reading many numbers in a table, a heatmap uses colors to show the strength of relationships.

### Creating a Correlation Matrix

```python
corr = df.corr(numeric_only=True)
corr
```

This calculates correlations between all numerical columns.

---

### Creating a Heatmap

```python
sns.heatmap(corr, annot=True)
plt.show()
```

### Parameters

```python
annot=True
```

Displays the correlation values inside each cell.

---

```python
cbar=False
```

Removes the color scale on the right side.

Example:

```python
sns.heatmap(corr, annot=True, cbar=False)
```

---

```python
cmap="coolwarm"
```

Changes the color scheme.

Example:

```python
sns.heatmap(corr, annot=True, cmap="coolwarm")
```

---

## Understanding the Heatmap

Each cell compares two features.

Example:

```text
Age ↔ Fare
```

The value inside the cell represents their correlation.

Example:

```text
0.17
```

Meaning:

> Age and Fare have a weak positive relationship.

---

## Why Rows and Columns Are the Same

Heatmaps compare every feature with every other feature.

Example:

```text
          Age   Fare   Survived

Age

Fare

Survived
```

Each cell answers:

> How strongly are these two features related?

Examples:

```text
Age ↔ Fare
Age ↔ Survived
Fare ↔ Survived
```

---

## Why the Diagonal Is Always 1

Examples:

```text
Age ↔ Age = 1
Fare ↔ Fare = 1
Survived ↔ Survived = 1
```

A feature is perfectly correlated with itself.

Therefore, the diagonal always contains:

```text
1.0
```

---

## How to Read a Heatmap

### Step 1

Identify the target feature.

Example:

```text
Target = survived
```

### Step 2

Look at the target row or column.

Example:

```text
Survived ↔ Age
Survived ↔ Fare
Survived ↔ Pclass
```

### Step 3

Find the strongest relationships.

Example:

```text
Survived ↔ Fare = 0.24
Survived ↔ Age = -0.05
```

Interpretation:

* Fare may be useful for prediction.
* Age has a very weak relationship with survival.

---

## Important Notes

* Color provides a quick visual clue.
* The number inside the cell is the actual correlation value.
* Always trust the correlation value more than the color.
