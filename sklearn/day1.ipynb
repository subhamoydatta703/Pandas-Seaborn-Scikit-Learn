### MAE — Mean Absolute Error
MAE does not fix the model.\
MAE only measures how wrong the model is.\
Lower MAE: Better

````python
from sklearn.metrics import mean_absolute_error
import numpy as np
````

````python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
````

````python
x = np.array([[1], [2], [3], [4], [5]])
y = np.array([30, 50, 100, 275, 495])

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)

model= LinearRegression()
model.fit(x_train, y_train)

prediction = model.predict(x_test)

mae = mean_absolute_error(y_test, prediction)
print("Actual:", y_test)
print("Predicted:", prediction)
print("MAE:", mae)

````

### MSE — Mean Squared Error

````python
from sklearn.metrics import mean_squared_error
````

````python
x = np.array([[1], [2], [3], [4], [5]])
y = np.array([30, 50, 100, 275, 495])

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)

model= LinearRegression()
model.fit(x_train, y_train)

prediction = model.predict(x_test)

mse = mean_squared_error(y_test, prediction)
print("Actual:", y_test)
print("Predicted:", prediction)
print("MSE:", mse)

````

````python
# Checking error
errors = abs(y_test - prediction)

print(errors)
````

````python
# Checking error
for actual, pred in zip(y_test, prediction):
    print(
        "Actual:", actual,
        "Predicted:", pred,
        "Error:", abs(actual - pred)
    )
````

### RMSE — Root Mean Square Error

MAE  → Average mistake\
MSE  → Average mistake with extra punishment for big errors\
RMSE → MSE made human-readable again

````python
x = np.array([[1], [2], [3], [4], [5]])
y = np.array([30, 50, 100, 275, 495])

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)

model= LinearRegression()
model.fit(x_train, y_train)

prediction = model.predict(x_test)

mse = mean_squared_error(y_test, prediction)
rmse = np.sqrt(mse)
print("Actual:", y_test)
print("Predicted:", prediction)
print("RMSE:", rmse)

````

# R² Score (R-Squared)

## What is R²?

R² measures how well the model learned the pattern/relationship in the data.

### Example

Study Hours → Marks

R² checks whether the model understood:

**More Study Hours → More Marks**

---

## R² Range

* **R² = 1.0** → Perfect model
* **R² ≈ 0.8–0.9** → Very good model
* **R² ≈ 0.5** → Average model
* **R² = 0** → Learned no useful pattern
* **R² < 0** → Very poor model

---

## Easy Memory

> **R² = How well did the model learn the pattern?**

---

# Error Metrics vs R²

## MAE (Mean Absolute Error)

### Question

How wrong are the predictions on average?

### Example

**MAE = 10**

Meaning:

The model is wrong by about **10 units on average**.

---

## MSE (Mean Squared Error)

### Question

Are there any large mistakes?

### Meaning

Large errors are punished heavily.

---

## RMSE (Root Mean Squared Error)

### Question

How wrong is the model while still penalizing large mistakes?

### Meaning

Same idea as MSE, but easier to understand because it returns to the original units.

---

## R² Score

### Question

How well did the model learn the relationship/pattern in the data?

---

# Simple Comparison

| Metric | Meaning                                  |
| ------ | ---------------------------------------- |
| MAE    | Average mistake                          |
| MSE    | Punishes large mistakes                  |
| RMSE   | Punishes large mistakes + original units |
| R²     | How well the model learned the pattern   |

---

# Important

### Error Metrics

Measure the **size of mistakes**.

### R²

Measures how well the model captures the **relationship between input and output**.

---

# Summary

* **MAE** → How wrong am I?
* **MSE** → How bad are the big mistakes?
* **RMSE** → How wrong am I (easy to understand)?
* **R²** → Did the model learn the pattern?

Use both **Error Metrics** and **R²** together when evaluating a regression model.


````python
from sklearn.metrics import mean_absolute_error, r2_score

# Hours studied
X = np.array([[1], [2], [3], [4], [5], [6]])

# Marks obtained
y = np.array([20, 35, 50, 65, 80, 95])

# Split data
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=42
)

# Train model
model = LinearRegression()
model.fit(X_train, y_train)

# Predictions
predictions = model.predict(X_test)

# Metrics
mae = mean_absolute_error(y_test, predictions)
r2 = r2_score(y_test, predictions)

print("X_test:", X_test.flatten())
print("Actual Marks:", y_test)
print("Predicted Marks:", predictions)

print("MAE:", mae)
print("R²:", r2)
````

# score() vs R² Score

## Are score() and R² the same?

For Linear Regression:

```python
model.score(X_test, y_test)
```

and

```python
r2_score(y_test, predictions)
```

return the same value.

Therefore:

**For LinearRegression, score() = R² Score**

---

## Why do people use r2_score() instead of score()?

### score()

```python
model.score(X_test, y_test)
```

* Short and convenient
* Automatically calculates the default metric of the model

---

### r2_score()

```python
r2_score(y_test, predictions)
```

* Explicit and clear
* Immediately tells readers that R² is being calculated
* Common in professional ML code

---

## Similarities

Both:

* Evaluate model performance
* Return the same value for Linear Regression
* Measure how well the model learned the pattern in the data

Example:

```python
score = model.score(X_test, y_test)

r2 = r2_score(y_test, predictions)
```

Output:

```text
0.92
0.92
```

---

## Differences

| score()               | r2_score()                 |
| --------------------- | -------------------------- |
| Method of the model   | Separate metric function   |
| Depends on model type | Always computes R²         |
| Shorter to write      | More explicit and readable |

---

## Important

For Linear Regression:

```text
score() = R² Score
```

For Classification Models:

```python
LogisticRegression().score(...)
```

returns:

```text
Accuracy
```

not R².

This is why many ML engineers prefer using:

```python
r2_score()
accuracy_score()
precision_score()
recall_score()
```

because the metric is explicitly visible in the code.

---

## Easy Memory

For Linear Regression:

```text
score() = shortcut for R²
```

For Classification:

```text
score() = usually Accuracy
```

Therefore, using metric functions like:

* r2_score()
* mean_absolute_error()
* mean_squared_error()
* accuracy_score()

makes the code easier to understand.
