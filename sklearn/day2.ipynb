# Classification Metrics

# Classification Labels

In classification, we predict **categories**, not numerical values.

### Examples

* Pass / Fail
* Spam / Not Spam
* Selected / Rejected
* Disease / No Disease

---

## Why Use 0 and 1?

Machine Learning models work best with numbers.

So categories are often represented as:

* Fail = 0
* Pass = 1

or

* Spam = 0
* Not Spam = 1

---

## Can We Use Strings?

Yes.

Example:

```python
y = ["Fail", "Fail", "Pass", "Pass"]
```

scikit-learn can handle many string labels directly.

If needed, scikit-learn internally converts these labels into numerical representations before training the model.

---

## Important

Even though we use numbers like 0 and 1, they are **labels**, not numerical quantities.

Example:

```python
[0, 1, 1, 0]
```

means:

```text
Fail, Pass, Pass, Fail
```

not numerical values.

---

## Key Difference

### Regression

Predicts numerical values:

* 85 Marks
* ₹50,000 Salary
* ₹20,00,000 House Price

### Classification

Predicts categories:

* Pass / Fail
* Spam / Not Spam
* Selected / Rejected

---

## Easy Memory

0 and 1 in classification are **category labels**, not actual numerical predictions.

If string labels are used, scikit-learn can internally convert them into numerical labels for training.


# Logistic Regression and Transformers

## Why Do We Need Transformers?

Machine Learning models work best with clean numerical data.

Transformers help prepare the input features (**X**) before training.

Common transformers:

* SimpleImputer → Handle missing values
* OneHotEncoder → Convert categorical features into numbers
* StandardScaler → Scale numerical features
* MinMaxScaler → Scale numerical features to a fixed range

---

## When Do We Need a Transformer?

### 1. Missing Values

Example:

| Age | Salary |
| --- | ------ |
| 20  | 30000  |
| 25  | NaN    |
| 30  | 50000  |

Use:

```python
SimpleImputer()
```

---

### 2. Categorical Features (Input X)

Example:

| Gender |
| ------ |
| Male   |
| Female |
| Male   |

Models cannot directly learn from text categories.

Use:

```python
OneHotEncoder()
```

Result:

| Male | Female |
| ---- | ------ |
| 1    | 0      |
| 0    | 1      |
| 1    | 0      |

---

### 3. Very Different Numerical Scales

Example:

| Age | Salary |
| --- | ------ |
| 20  | 20000  |
| 40  | 200000 |

Use:

```python
StandardScaler()
```

or

```python
MinMaxScaler()
```

---

## Important: Target (y) vs Features (X)

### Target (y)

Example:

```python
y = ["Pass", "Fail", "Pass"]
```

scikit-learn can often internally convert labels such as:

```text
Pass → 1
Fail → 0
```

because they are target classes.

---

### Features (X)

Example:

```python
X = [["Male"], ["Female"], ["Male"]]
```

If we directly convert:

```text
Male → 0
Female → 1
```

the model may incorrectly think:

```text
Female > Male
```

which is meaningless.

Therefore we use:

```python
OneHotEncoder()
```

for categorical input features.

---

## Why Not Use Only Transformers?

Transformers prepare data but do not learn patterns.

They cannot answer:

```text
Hours Studied = 4

Pass or Fail?
```

Transformers have:

```python
fit()
transform()
fit_transform()
```

but no prediction capability.

---

## Why Use Logistic Regression?

Logistic Regression learns the relationship between features (X) and target (y).

Example:

```text
Hours Studied → Pass / Fail
```

The model learns patterns such as:

```text
More Study Hours
        ↓
Higher Chance of Passing
```

and can make predictions for new data.

---

## Workflow

### If Data Needs Transformation

```text
Raw Data
    ↓
Need Transformation?
    ↓
Yes
    ↓
Transformer
    ↓
Clean Numerical Data
    ↓
Logistic Regression
    ↓
Learn Patterns
    ↓
Predictions
```

### If Data Is Already Clean and Numerical

```text
Raw Numerical Data
        ↓
Logistic Regression
        ↓
Learn Patterns
        ↓
Predictions
```

---

## Example

No transformer needed:

```python
X = [[1], [2], [3], [4], [5]]
```

because:

* Numerical ✔
* No missing values ✔
* No categorical features ✔

---

## Easy Memory

```text
Transformer
↓
Prepare the input data (X)

Logistic Regression
↓
Learn patterns from X and y

Predict
↓
Pass / Fail
```

---

## Key Idea

Transformers are used only when the input features (X) need preparation.

Logistic Regression is used to learn patterns and make predictions.

We do not use transformers instead of Logistic Regression.

We use transformers first (if needed), then train Logistic Regression.


````python
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
````

````python
# Hours studied
x = np.array([
    [1],
    [2],
    [3],
    [4],
    [5],
    [6]
])

# Result: Pass or Fail
y = np.array([
    "Fail",
    "Fail",
    "Fail",
    "Pass",
    "Pass",
    "Pass"
])

x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.33, random_state=42
)

model = LogisticRegression()
model.fit(x_train, y_train)
predictedValue = model.predict(x_test)
print(predictedValue)
print(y_test)
predictedValue = model.predict([[0.5]])
print(predictedValue[0])

````

````python
print(model.classes_)
print(model.predict_proba([[4]])*100)
# 29.093 % chance of fail and 70.906 % chance of pass
````

````python
print(model.score(x_test, y_test))
````
