# ColumnTransformer

## Definition

`ColumnTransformer` allows us to apply different transformers to different columns of a dataset.

---

## Why Use It?

Different columns often need different preprocessing.

Example:

```text
Gender
↓
OneHotEncoder

Age, Salary
↓
StandardScaler
```

Instead of applying transformers manually and combining the results, `ColumnTransformer` does everything automatically.

---

## Structure

```python
ColumnTransformer(
    transformers=[
        (
            "name",
            transformer(),
            ["column1", "column2"]
        )
    ]
)
```

### Parts

```text
(
    Name,
    Transformer,
    Columns
)
```

* **Name** → Label required by sklearn.
* **Transformer** → Operation to perform.
* **Columns** → Columns on which the operation is applied.

---

## Example

```python
preprocessor = ColumnTransformer(
    transformers=[
        (
            "gender_encoder",
            OneHotEncoder(),
            ["Gender"]
        ),
        (
            "number_scaler",
            StandardScaler(),
            ["Age", "Salary"]
        )
    ]
)
```

Meaning:

```text
Gender
↓
OneHotEncoder

Age, Salary
↓
StandardScaler
```

---

## Flow

```text
Dataset
    ↓
ColumnTransformer
    ↓
Different transformers
on different columns
    ↓
Transformed Dataset
```

---

## Memory

```text
ColumnTransformer
↓
Apply different transformers
to different columns automatically.
```


````python
import pandas as pd

df = pd.DataFrame({
    "Age": [20, 30, 40],
    "Gender": ["Male", "Female", "Male"],
    "Salary": [20000, 40000, 60000]
})

print(df)
````

````python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
````

````python
preprocessor = ColumnTransformer(
    transformers=[
        (
            "gender_encoder",
            OneHotEncoder(),
            ["Gender"]
        ),
        (
            "numerical_encoder",
            StandardScaler(),
            ["Age","Salary"]
        )
    ]
)
````

````python
transformed_data = preprocessor.fit_transform(df)
print(transformed_data)
````

````python
for org_data, trans_data in zip(df.values, transformed_data):
    print(f"Original data: {org_data}")
    print(f"Transformed data: {transformed_data}")
    print()
````

````python
for row in transformed_data:
    female, male, age_scaled, salary_scaled = row

    print(
        f"Female={female}, "
        f"Male={male}, "
        f"Age_scaled={age_scaled:.2f}, "
        f"Salary_scaled={salary_scaled:.2f}"
    )
````
