## models, test, predict

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
model = LinearRegression()
```

```python
X = np.array([[1], [2], [3], [4], [5]])
y = np.array([30, 40, 50, 60, 70])


X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)
# print(X_test)
# print(y_test)
model.fit(X_train, y_train)
predictVal = model.predict(X_test)

predictVal = np.clip(predictVal, 0, 100)
# setting range of prediction value of 0-100 => cause maximum marks is 100 and minimum marks is 0
```

```python
print(predictVal)
print()
print(y_test)
```

#### Score checks the performance of the trained model on the test data.

0.0  → completely wrong\
0.5  → 50% correct\
0.8  → 80% correct\
1.0  → 100% correct

```python
print(model.predict([[18]]))
print(model.score(X_test, y_test))
```

#### `model.fit()` is not just memorizing it is also for continuous error correction and optimization

### Transformers, transform(), and fit_transform()

#### StandardScaler --> Transformer

```python
from sklearn.preprocessing import StandardScaler
# get tool or transformer
```

```python
X = np.array([
    [500000],
    [900000],
    [300000]
])

scaler = StandardScaler() #call tool or transformer
val = scaler.fit_transform(X) #study & transform(scaling) the data
```

```python
print(val)
```

#### OneHotEncoder --> Transformer

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
X = np.array([
    ["Kolkata"],
    ["Delhi"],
    ["Mumbai"]
])

scaler = OneHotEncoder()
res = scaler.fit_transform(X)
```

```python
print(res.toarray()) # converting sparse matrix(res) into array(for learning purpose only)
#but in real cases we did not convert it into array and use that sparse matrix
```

```python
print(res)

# (0, 1)    1.0

# means:
# row 0
# column 1
# value = 1
```

## SimpleImputer (Handling Missing Data)
#### SimpleImputer is a preprocessing tool in scikit-learn that: fills missing values using a chosen strategy.

```python
from sklearn.impute import SimpleImputer
```

```python
X = np.array([
    [20],
    [np.nan],
    [30]
])

imputer = SimpleImputer(strategy="mean")
filled = imputer.fit_transform(X)

print(filled)
```
